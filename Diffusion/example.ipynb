```python
# Cell 1 - the tokenizer: encode a sequence onto the canvas, then decode it back
from config import TokenizerConfig
from tokenizer import AAVTokenizer

tokenizer = AAVTokenizer(TokenizerConfig())

seq = "ADEEIRTTNPVATEQYGSVSTNPvQRGNR"  # uppercase = the 28 anchors, lowercase v = an insertion
canvas = tokenizer.encode(seq)

print("sequence:    ", seq)
print("vocab size:  ", tokenizer.vocab_size)
print("canvas length:", tokenizer.canvas_len)
print("canvas ids:  ", canvas)
print("decoded back:", tokenizer.decode(canvas))
```

```python
# Cell 2 - the dataloader: load the pre-tokenized dataset and look at one batch
from DataLoading import get_dataloader

loader = get_dataloader(batch_size=128)

print("dataset size:", len(loader.dataset))
print("num batches: ", len(loader))

canvas_batch, score_batch = next(iter(loader))
print("canvas batch shape:", tuple(canvas_batch.shape), canvas_batch.dtype)
print("score batch shape: ", tuple(score_batch.shape), score_batch.dtype)
```
